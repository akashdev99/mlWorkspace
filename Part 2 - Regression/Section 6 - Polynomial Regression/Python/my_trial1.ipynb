Import Libraries

```python
import numpy as np
import pandas as pd
```

Immport dataset``

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[: , 1:-1].values
y = data.iloc[: , -1].values
```

Create X and Y 

```python
print(X)
print(y)
```

Make Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X , y)
```

Pridict with Linear Regression

```python
y_pred=lr.predict(X)
```

Visualize prediction With Linear Regression


```python
import matplotlib.pyplot as pl


pl.scatter(X , y, color = "red")
pl.plot(X , y_pred , color = "blue")
pl.title("Test With Linear Regression Predictions")
pl.show()
```

Performance - Linear Regression

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

print("MSE=",mean_squared_error(y_true = y ,y_pred= y_pred))
print("MAE=",mean_squared_error(y_true = y ,y_pred= y_pred))
print("R2 Score=",r2_score(y_true = y ,y_pred= y_pred))
```

Polynomial Feature Addition


```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg  = PolynomialFeatures(degree=4)

X_poly = poly_reg.fit_transform(X)
print(X_poly)

lr2 = LinearRegression()
lr2.fit(X_poly , y)
```

```python
y_pred_2=lr2.predict(X_poly)
```

```python
import matplotlib.pyplot as pl


pl.scatter(X , y, color = "red")
pl.plot(X , y_pred_2 , color = "blue")
pl.title("Test With Linear Regression Predictions")
pl.show()
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

print("MSE=",mean_squared_error(y_true = y ,y_pred= y_pred_2))
print("MAE=",mean_squared_error(y_true = y ,y_pred= y_pred_2))
print("R2 Score=",r2_score(y_true = y ,y_pred= y_pred_2))
```
